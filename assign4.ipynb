```python
def func(x):
  return 2*(x+3)
```

```python

def run(start,n_iteration,learning_rate):
  x=start
  gra=[]
  for i in range(n_iteration):
    grad=func(x)
    x=x-learning_rate*grad
    gra.append(x)
  return x,gra
```

```python
start = 2
n_iteration = 50
learning_rate = 0.1
min_x,gra=run(start,n_iteration,learning_rate)
print(min_x)
```

```python
import seaborn as sns
sns.scatterplot(x=range(len(gra)),y=gra)
```
